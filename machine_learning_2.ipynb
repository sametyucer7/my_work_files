```
import pandas as pd
from sklearn.linear_model import LinearRegression
```



**Preprocessing ve Pandas Dummy Metodu**



```
df = pd.read_csv("arabalist.csv")
df.head(3)
```

```
# bu listedeki arabaların fiyat denklemini yapmak için önce nelerin araba fiyatı üstünde

# etkiye sahip olduğunu ve olmadığına bakıp olanları ve olmayanları ayırmalıyız

# Href column daki gibi arabanın satıldığı yerdeki linki arabanın fiyatını etkilemez
```

```
df = df.drop(columns=["index" , "href" , "MileageRank" , "PriceRank" , "PPYRank" , "Score"])
```

```
df.head(3)

# Type gibi katogorik değerler yani işleme giremicek değerlerdir . (fuel , transmission , engine)
# * engine rakamların yanında L yazdığı için numerik olmaktan çıkar
```

```
df.columns = ["Yıl" , "Tür" , "Mil" , "Motor" , "Ps" , "Vites" , "Yakıt" , "Sahip" , "Fiyat" , "Ppy"]
df.head(3)
```

```
df["Motor"] = df["Motor"].str.replace("L", "")
df["Motor"] = pd.to_numeric(df["Motor"])
```

```
df = pd.get_dummies(df , columns = ["Tür" , "Vites" , "Yakıt"]  , dtype=int  , drop_first=True)
```

```
df.head(3)
```

```
y = df["Fiyat"]
x = df.drop("Fiyat" , axis = 1)
```

```
lm = LinearRegression()
model = lm.fit(x , y)
```

```
model.predict([[2017 , 30000 , 1.6, 110 , 1 , 2600 , 0 , 1]])
```

```
model.score(x,y) * 100
```

**----- Makine Öğrenmesinde Train Test Split Metodu**

```
from sklearn.model_selection import train_test_split
```

```
df = pd.read_csv("arabalist.csv")
```

```
df = df.drop(columns=["index" , "href" , "MileageRank" , "PriceRank" , "PPYRank" , "Score"])
```

```
df.columns = ["Yıl" , "Tür" , "Mil" , "Motor" , "Ps" , "Vites" , "Yakıt" , "Sahip" , "Fiyat" , "Ppy"]
```

```
df["Motor"] = df["Motor"].str.replace("L", "")
df["Motor"] = pd.to_numeric(df["Motor"])
```

```
df = pd.get_dummies(df , columns = ["Tür" , "Vites" , "Yakıt"]  , dtype=int  , drop_first=True)
```

```
y = df["Fiyat"]
x = df.drop("Fiyat" , axis = 1)
```

```
# verilerin 100 de 100 ünü öğretmeyip belirlenen bir kısmı öğretip geri kalanını
# modelin hiç görmediği bu veriler üzerinde test ederiz.
# Böylece modelin sadece ezber yapıp yapmadığını değil,
# gerçekten genelleme yapıp yapamadığını ölçeriz.
```

```
x_train , x_test , y_train ,y_test = train_test_split(x , y , train_size=70 , random_state=22)
```

```
lm = LinearRegression()
model = lm.fit(x_train , y_train)
model.score(x_test , y_test)
```
