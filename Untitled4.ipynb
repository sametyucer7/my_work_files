```python
##################################################################################################################

###################    DATA FRAMELERÄ° BÄ°RLEÅžTÄ°RME      ###########################################################

Ä°NNER = Ä°KÄ° DATA FRAMEÄ°N ORTAK DATALARI

OUTER = Ä°KÄ° DATA FRAMEDEKÄ° ORTAK OLAMAYAN KENDÄ°LERÄ°NE Ã–ZGÃœ VERÄ°LER
```

```python
df1= pd.DataFrame(columns=["fellowshiplD" , "Firstname" , "Skills"])
new_rows = [{"fellowshiplD" : "1001" , "Firstname" : "Frodo" ,  "Skills" : "Hiding"}
            ,{"fellowshiplD" : "1002" , "Firstname" : "Samwise" ,  "Skills" : "Gardening"}
            ,{"fellowshiplD" : "1003" , "Firstname" : "Gandalf" ,  "Skills" : "Spelss"}
            ,{"fellowshiplD" : "1004" , "Firstname" : "Pippin" ,  "Skills" : "Fireworkks"}]

df1= pd.concat([df1, pd.DataFrame(new_rows)],ignore_index=True)
```

```python
df2= pd.DataFrame(columns=["fellowshiplD" , "Firstname" , "Age"])

new_rovs = [{"fellowshiplD" : "1001" , "Firstname" : "Frodo" ,  "Age" : "50"}
            ,{"fellowshiplD" : "1002" , "Firstname" : "Samwise" ,  "Age" : "39"}
            ,{"fellowshiplD" : "1006" , "Firstname" : "Legolas" ,  "Age" : "2931"}
            ,{"fellowshiplD" : "1007" , "Firstname" : "Elrond" ,  "Age" : "6520"}
             ,{"fellowshiplD" : "1008" , "Firstname" : "Boromir" ,  "Age" : "51"}]
df2= pd.concat([df2, pd.DataFrame(new_rovs)],ignore_index=True)
```

```python
df1.merge(df2 , how="inner") # Ä°KÄ° DATA FRAMEÄ°N ORTAKLARIIINI BÄ°RLEÅžTÄ°RÄ°R VE BASE HALÄ° Ä°NNERDIR


df1.merge(df2 , how="outer") # OUTER OLUNCA VEYA ANLAMI GÄ°BÄ° ORTAK OOLANLARI BÄ°RLEÅžTÄ°RÄ°R AMA TABLOYA OLMAYANLARIDA ALIR

##############       DF1 ===== LEFT DATAFRAME

##############       DF2 ===== RÄ°GHT DATAFRAME 

#############        DATA FRAMEDEKÄ° NANLARI FÄ°LL.NA() Ä°LE DOLDURABÄ°LÄ°RÄ°Z


df1.merge(df2 , how = "left" )       DF1 Ä° SABÄ°T TUTTU VE DF1 DE OLAN DEÄžERLER Ä°Ã‡Ä°N DF2 DE VARSA ORDAKÄ° SÃœTUN BAÅžLIKLARINI VE DEÄžERLERÄ°NÄ°DE EKLEDÄ°
```

```python
df1.merge(df2 , how = "cross" ).head(3) ############## DATA F 1 DEKÄ° TÃœM DEÄžERLER DF2 DEKÄ° DEÄžERLER Ä°LE KARÅžILAÅžTIRIYOR  
```

```python
pd.concat([ df1 , df2 ], join = "inner") ########## Ä°NNER JOÄ°N OLUNCA ORTAKK BAÅžLIKLARI ALDI

                                         ######### OURER OLUNCA TÃœM BAÅžLIKLARI ALIR

                                         #########
```
