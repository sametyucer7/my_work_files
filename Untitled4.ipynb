```python
##################################################################################################################

###################    DATA FRAMELERİ BİRLEŞTİRME      ###########################################################

İNNER = İKİ DATA FRAMEİN ORTAK DATALARI

OUTER = İKİ DATA FRAMEDEKİ ORTAK OLAMAYAN KENDİLERİNE ÖZGÜ VERİLER
```

```python
df1= pd.DataFrame(columns=["fellowshiplD" , "Firstname" , "Skills"])
new_rows = [{"fellowshiplD" : "1001" , "Firstname" : "Frodo" ,  "Skills" : "Hiding"}
            ,{"fellowshiplD" : "1002" , "Firstname" : "Samwise" ,  "Skills" : "Gardening"}
            ,{"fellowshiplD" : "1003" , "Firstname" : "Gandalf" ,  "Skills" : "Spelss"}
            ,{"fellowshiplD" : "1004" , "Firstname" : "Pippin" ,  "Skills" : "Fireworkks"}]

df1= pd.concat([df1, pd.DataFrame(new_rows)],ignore_index=True)
```

```python
df2= pd.DataFrame(columns=["fellowshiplD" , "Firstname" , "Age"])

new_rovs = [{"fellowshiplD" : "1001" , "Firstname" : "Frodo" ,  "Age" : "50"}
            ,{"fellowshiplD" : "1002" , "Firstname" : "Samwise" ,  "Age" : "39"}
            ,{"fellowshiplD" : "1006" , "Firstname" : "Legolas" ,  "Age" : "2931"}
            ,{"fellowshiplD" : "1007" , "Firstname" : "Elrond" ,  "Age" : "6520"}
             ,{"fellowshiplD" : "1008" , "Firstname" : "Boromir" ,  "Age" : "51"}]
df2= pd.concat([df2, pd.DataFrame(new_rovs)],ignore_index=True)
```

```python
df1.merge(df2 , how="inner") # İKİ DATA FRAMEİN ORTAKLARIIINI BİRLEŞTİRİR VE BASE HALİ İNNERDIR


df1.merge(df2 , how="outer") # OUTER OLUNCA VEYA ANLAMI GİBİ ORTAK OOLANLARI BİRLEŞTİRİR AMA TABLOYA OLMAYANLARIDA ALIR

##############       DF1 ===== LEFT DATAFRAME

##############       DF2 ===== RİGHT DATAFRAME 

#############        DATA FRAMEDEKİ NANLARI FİLL.NA() İLE DOLDURABİLİRİZ


df1.merge(df2 , how = "left" )       DF1 İ SABİT TUTTU VE DF1 DE OLAN DEĞERLER İÇİN DF2 DE VARSA ORDAKİ SÜTUN BAŞLIKLARINI VE DEĞERLERİNİDE EKLEDİ
```

```python
df1.merge(df2 , how = "cross" ).head(3) ############## DATA F 1 DEKİ TÜM DEĞERLER DF2 DEKİ DEĞERLER İLE KARŞILAŞTIRIYOR  
```

```python
pd.concat([ df1 , df2 ], join = "inner") ########## İNNER JOİN OLUNCA ORTAKK BAŞLIKLARI ALDI

                                         ######### OURER OLUNCA TÜM BAŞLIKLARI ALIR

                                         #########
```
